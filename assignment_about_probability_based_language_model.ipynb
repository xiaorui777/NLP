### 2. What's the Probability Model?

Ans：
- 用来描述不同随机变量之间关系的数学模型，通常情况下刻画了一个或多个随机变量之间的相互非确定性的概率关系。

- 从数学上讲，该模型通常被表达为
(Y,P)其中Y是观测集合用来描述可能的观测结果，P是Y对应的概率分布函数集合。若使用概率模型，一般而言需假设存在一个确定的分布P生成观测数据Y。因此通常使用统计推断的办法确定集合P中谁是数据产生的原因。
大多数统计检验都可以被理解为一种概率模型。

### 5. What's the Language Model;

Ans：
- 简单地说，语言模型就是用来计算一个句子的概率的模型，也就是判断一句话是否是人话的概率。合法的句子得到概率比较大，而不合法的句子得到的概率比较小，这样只需要从所有可能的情况中选取概率最大的那种组合，我们就能够得到合法的句子了。

### 7. What's the 1-gram language model;

Ans：
- 把每个词出现的概率看成是独立事件，一句话的联合概率即每一个词出现的概率的乘积

$$P(sentence) = P（word1)*P(word2)*...*P(wordn)$$


### 8. What's the disadvantages and advantages of 1-gram language model;

Ans：
- 由于1-gram只考虑单个单词的概率，那么缺点就是如果一个句子越长，这个句子的联合概率就越低
- 没有考虑到单词之间的关系和联系

### 9. What't the 2-gram models;

Ans:
- 引入条件概率，即前一个词出现的情况下，后一个词出现的概率为多少，即考虑到相邻两个词之间的关系
$$P(sentence) = P（word1)*P(word2|word1)*P(word3|word2)*...*P(wordn|wordn-1)$$

```
files = glob.glob(file_path)

for name in files:
    try:
        text_file = open(name, "r", encoding="utf8")
        lines = text_file.readlines()
        all_articles.append(lines)
    except IOError as exc: # not sure what error this is
        if exc.errno != errno.EISDIR:
            raise
```

```
# 将文章正则化，把'\n','<doc id','</doc>'删除
Text = ''

for articles in all_articles:
    for i in articles:
        if i.startswith('<doc id') or i.startswith('</doc>'):
            articles.remove(i)
        else:
            Text += i
Text = Text.replace('\n',' ')
```

### 实现语言模型（1-gram model）

```
# Step3:实现语言模型 （v1: the 1_gram model）
```

```
# 得到每个词出现的频率
from collections import Counter
words_count = Counter(valid_tokens)
words_count.most_common(10)
```

```
# 将频率图画出来
import matplotlib.pyplot as plt
%matplotlib inline

frequences = [f for w, f in words_count.most_common(200)]
x = [i for i in range(len(frequences))]
```

```
plt.plot(x,frequences)
```

```
frequences_all = [f for w,f in words_count.most_common()]
frequences_sum = sum(frequences_all)
frequences_sum
```

```
# word出现的概率
def get_prob(word):
    return words_count[word]/frequences_sum
```

```
get_prob('玉米')
```

```
# 定义一个连乘函数
from functools import reduce
def produce(numbers):
    return reduce(lambda n1,n2:n1*n2,numbers)
```

```
# 1-gram model
def language_model_one_gram(string):
    words = cut(string)
    return produce([get_prob(w) for w in words])
```

```
language_model_one_gram('宇宙中观测到了新的行星')
```

```
language_model_one_gram('安卓操作系统')
```

### 改进语言模型（2-gram model）

```
all_2_grams_words = [''.join(valid_tokens[i:i+2]) for i in range(len(valid_tokens[:-2]))]
_2_gram_sum = len(all_2_grams_words)
_2_gram_counter = Counter(all_2_grams_words)
```

```
# 求 word1 和 word2 的联合概率，即 p(word1*word2)
def get_combination_prob(w1,w2):
    if w1+w2 in _2_gram_counter:
        return _2_gram_counter[w1+w2]/_2_gram_sum
    else:
        return 1/_2_gram_sum
```

```
# 求 p(word1|word2)即 p(word1*word2)/p(word1)
def get_prob_2_gram(w1,w2):
    return get_combination_prob(w1,w2)/get_prob(w1) #1-gram概率
```

```
get_prob_2_gram('吃','火锅')
```

```
# 2-gram model
def language_model_two_gram(sentence):
    sentence_probability = 1
    words = cut(sentence)
    for i,word in enumerate(words):
        if i == 0:
            prob = get_prob(word)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous,word)
        sentence_probability *= prob
    return sentence_probability
```

```
language_model_two_gram('小明说今晚去吃火锅')
```

```
language_model_two_gram('天方夜谭也许你说是的')
```
